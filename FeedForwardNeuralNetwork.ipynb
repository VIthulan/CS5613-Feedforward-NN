```python

data_df['A1_up'] = data_df["A1"].replace({"?": "b"})
data_df['A4_up'] = data_df["A4"].replace({"?": "u"})
data_df['A5_up'] = data_df["A5"].replace({"?": "g"})
data_df['A6_up'] = data_df["A6"].replace({"?": "c"})
data_df['A7_up'] = data_df["A7"].replace({"?": "v"})
data_df['A14_up'] = data_df["A14"].replace({"?": "00000"})
data_df['A2_up'] = data_df["A2"].replace({"?": 31.02})
```

```python
data_df['A2_up'] = data_df["A2_up"].astype(float)
```

```python
labelencoder = LabelEncoder()

data_df['A1_encd'] = labelencoder.fit_transform(data_df['A1_up'])
data_df['A4_encd'] = labelencoder.fit_transform(data_df['A4_up'])
data_df['A5_encd'] = labelencoder.fit_transform(data_df['A5_up'])
data_df['A6_encd'] = labelencoder.fit_transform(data_df['A6'])
data_df['A7_encd'] = labelencoder.fit_transform(data_df['A7_up'])
data_df['A9_encd'] = labelencoder.fit_transform(data_df['A9'])
data_df['A10_encd'] = labelencoder.fit_transform(data_df['A10'])
data_df['A12_encd'] = labelencoder.fit_transform(data_df['A12'])
data_df['A13_encd'] = labelencoder.fit_transform(data_df['A13'])
data_df['A14_encd'] = labelencoder.fit_transform(data_df['A14_up'])
data_df['A16_encd'] = labelencoder.fit_transform(data_df['A16'])
```

```python
feature_df = data_df[['A1_encd', 'A2_up', 'A3', 'A4_encd', 'A5_encd', 'A6_encd', 'A7_encd', 'A8', 'A9_encd', 'A10_encd', 'A11', 'A12_encd', 'A13_encd', 'A14_encd', 'A15', 'A16_encd']]
```

```python
for i in range(0,16):
    feature_df = feature_df.rename(columns={feature_df.columns[i]: 'A%s'%(i+1)})

feature_df = feature_df.rename(columns={feature_df.columns[15]: 'output'})
```

```python
feature_df.describe()
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
from sklearn.model_selection import KFold
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix,classification_report


# from keras.models import Sequential
# from keras.layers import Dense
```

```python
feat_arr = []
for i in range (1,16):
    feat_arr.append('A%s'%i)
```

```python
feat_arr
```

```python
X = feature_df[feat_arr].to_numpy()
Y = feature_df.output.to_numpy()
```

```python
def get_model():
    model = Sequential()
    model.add(Dense(10, input_dim=15, activation='relu', activity_regularizer='l1_l2'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
# model.fit(feature_df[feat_arr], feature_df.output, epochs=150, batch_size=10)
```

```python
kf = KFold(n_splits = 5, shuffle = True, random_state = 2)


# result = next(kf.split(feature_df), None)
```

```python
# Y[result[0]]
```

```python
cvscores = []
f1scores = []

kff = 1

kf = KFold(n_splits = 5, shuffle = True, random_state = 2)

for train_index, test_index in kf.split(feature_df):
    model = get_model()
    model.fit(X[train_index], Y[train_index], epochs=150, batch_size=10, verbose=0)
    score = model.evaluate(X[test_index], Y[test_index], verbose=0)

    print("Fold %s -- %s: %.2f%%" % (kff, model.metrics_names[1], score[1]*100))
    kff = kff + 1

    cvscores.append(score[1])
    
print('\n-------- Overall results ----')
print("F1 %.4f%% (+/- %.4f%%)" % (np.mean(cvscores), np.std(cvscores)))
# print("F1 Score %.2f%% (+/- %.2f%)" % (np.mean(f1scores), np.std(f1scores)))
```
